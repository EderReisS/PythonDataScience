```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('input_Files/agrupamento.csv')
```

```python
df.describe
```

## Entrada de dados

```python
centers = pd.read_csv('input_Files/centroides_iniciais.csv')
```

```python
centers.describe
```

## Visualização

```python
sns.relplot(data = df )
```

```python
sns.pairplot(df)
```

## Dados de treinamento

```python
X = np.array(df)
```

```python
X
```

## K-means

### Clusters de tamanho n = 4

#### Definição do modelo

```python
kmeans_4 = KMeans(n_clusters=4 , random_state=0)
```

```python
kmeans_4.fit(X)
```

```python
kmeans_4.labels_
```

#### Aplicação nos dados

```python
df_A4 = df.copy()
```

```python
df_A4['CLUSTER']= kmeans_4.labels_
```

#### Visualização dos dados

```python
sns.pairplot(df_A4, hue='CLUSTER', palette ='dark')
```

### Clusters de tamanho n = 12

#### Definição do modelo

```python
kmeans_12 = KMeans(n_clusters=12 , random_state=0)
```

```python
kmeans_12.fit(X)
```

```python
kmeans_12.labels_
```

#### Aplicação nos dados

```python
df_A12 = df.copy()
```

```python
df_A12['CLUSTER']= kmeans_12.labels_
```

#### Visualização dos dados

```python

sns.pairplot(df_A12, hue='CLUSTER')
```

# Dado o salário fixo, o valor das vendas efetuadas pelo vendedor
Onde, sabendo-se que ele recebe uma comissão de 3% sobre o total das vendas até R$ 1.500,00 e 5% sobre o que ultrapassar este valor, calcular e escrever o seu salário total.

```python

salario = 0
comissao=0
valor_ref=1500.0

periodo = True

while periodo:
    entrada=input("Por favor, valor de 'venda' para inserir o valor da venda ou 'fim' para determinação do salário no final do período: ")
    
    if (entrada=='venda'):
        venda=float(input("Venda em R$: "))
        
        if (venda<=valor_ref):
            salario=salario+round(venda*0.03,2)
        else:
            salario=salario+round(venda*0.05,2)
            
    elif (entrada=='fim'):
        periodo=False #fim do periodo
    
    else:
        print("\n Por favor digite 'venda' ou 'fim'")
        


print(f"\nO salário no fim do periodo é {salario:0.2f}!")
    

    
```

# Dado um número, imprima na tela todos os seus divisores

```python
N=0

MSG="Determinação dos divisores do número n, por vaor digite o número 'n' \n"

MSG_alert_2="\nPor favor, digite um número positivo\n"
MSG_alert_1="\nPor favor, digite um número inteiro válido \n"
MSG_alert_0="Pode-se dividir nada(ou zero) para quaisquer n número!!"

N_valido=False


def divisores(n):
    
    for i in range(1,n+1):
        if ((n%i)==0):
            print(f"{i} é divisor de {n}")
        
    return None

while (not N_valido):
    try:
        N=int(input(MSG))
        if (N<0):
            print(MSG_alert_2)
        else:
            N_valido=True
        
    except:
        print(MSG_alert_1)
        
if (N==None or N=='' or N==0):
    print(MSG_alert_0)
else:
    divisores(N)
```

# Calcule a soma de mil termos dos inversos dos fatoriais:
1/(1!) + 1/(2!) + 1/(3!) + ...

```python
total=0
n=1000

def fatorial(n):
    fatorial=1
    for i in range(n+1):
        if i==0:
            i=1
            
        fatorial=fatorial*i
        
    return fatorial

for i in range(1,n+1):
    total=total+1/fatorial(i)

print(f"\n {n} primeiros termos: \n\n {total:0.200f}")
```

# Faça um algoritmo para simular uma situação simples de depósito, retirada e consulta em um banco. O algoritmo inicialmente deverá mostrar um menu com as seguintes opções:
- 1 - Depósito
- 2 - Retirada
- 3 - Saldo
- 4-  Sair do algoritmo
<br> Se a escolha do usuário for depósito ou retirada, o algoritmo deverá pedir o valor da operação e atualizar automaticamente o valor existente na conta.
O algoritmo deverá ser utilizado até que o usuário escolha a opção sair do algoritmo.

```python
SessaoAtiva=True

OpcaoValida=False


SALDO =4200.13
VALOR=0


def valor(msg,msg_alert):
    valor_valido=False
    
    while (not valor_valido):
        try:
            valor=float(input(msg))
            if (valor<0): #valores de depósito/retirada devem ser positivos
                print(msg_alert)
            else:
                valor_valido=True
                return valor
            
        except:
            print(msg_alert)
        

def dep(valor,saldo):
    valor=round(valor,2)
    saldo=valor+saldo
    
    print(f"\nDepósito no valor de R$ {valor:0.2f} realizado com sucesso.\n")
    return saldo

def ret(valor,saldo):
    valor=round(valor,2)
    saldo=saldo-valor
    
    if saldo<0:
        print("\nRetirada não autorizada: saldo insuficiente.\n")
        saldo=saldo+valor
        return saldo
    else:
        print(f"\nRetirada no valor de R$ {valor:0.2f} realizada com sucesso.\n")
        return saldo
    
    


MSG_1='''\n Opções deste terminal:
1 - Depósito
2 - Retirada
3 - Saldo
4- Sair do algoritmo'''

MSG_2="Digite uma das opçoẽs: "

MSG_fim="\n\n\n*Sessão no terminal finalizada, obrigado!"

MSG_alerta_op="\n Por favor digite uma opção válida \n"


MSG_alert_v="Por favor digite uma um valor de operação válido!!\n"
MSG_0="*Operação cancelada."

#Deposito
MSG_dep="Informe o valor de depósito: \n"

#Retirada
MSG_ret="Informe o valor de retirada: \n"
MSG_saldo_0="*Operação indisponível pois o saldo está zerado"




while SessaoAtiva:
    print(MSG_1)
    while (not OpcaoValida):
        entrada=input(MSG_2)        
        OpcaoValida=(entrada=='1')or(entrada=='2')or(entrada=='3')or(entrada=='4')
        
        if (not OpcaoValida):
            print(MSG_alerta_op)
            
        
    if (entrada =='1'):
        VALOR=valor(MSG_dep,MSG_alert_v)
        if VALOR==0:#volta a tela inicial de opções caso não seja informado nenhum valor
            print(MSG_0)
            OpcaoValida=False
        else:
            SALDO=dep( VALOR, SALDO)
            OpcaoValida=False
            
        
        
        
    elif (entrada =='2'):
        
        
        if (not SALDO):
            print(MSG_saldo_0)
            OpcaoValida=False
            
        else:
            VALOR=valor(MSG_ret,MSG_alert_v)
            
            if VALOR==0:#volta a tela inicial de opções caso não seja informado nenhum valor
                print(MSG_0)
                OpcaoValida=False
            
            else:
                SALDO=ret( VALOR, SALDO)
                OpcaoValida=False        
        
    elif (entrada =='3'):
        print(f"\n \nO valor de saldo na conta é R$ {SALDO:0.2f}.\n")
        OpcaoValida=False
        
    else:
        SessaoAtiva=False
        
        
print(MSG_fim)

```
